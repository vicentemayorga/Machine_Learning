## 3.4 Filtros

Filtro de una columna:
- dataframe[dataframe["column"]==valor] #igual a valor
- dataframe[dataframe["column"]!=valor] #diferente a valor
- dataframe[dataframe["coliumn"].isin(lista_valores)] #Está contenido en una lista de valores

Filtro de 2 o más columnas
- dataframe[(dataframe["column1"]==valor_1)&(dataframe["column2"]==valor_2)] #Se tiene que cumplir condución 1 y 2
- dataframe[(dataframe["column1"]==valor_1)|(dataframe["column2"]==valor_2)] #Se puede cumplir condición 1 o 2

```python
df[df["Gender"].isin(["Female","Male", "otro_1", "otro_1"])]
```

```python
df[df["Region"]=="Europe"]
```

```python
df[(df["Region"]=="Europe")&(df["Gender"]=="Female")] #dataframe[ (condicion1) & (condicion2) & () ..]
```

# 4 Missing values
Datos que no fueron observados en el proceso de registro del dataset a utilizar. El mecanismo de generación de datos faltantes podría ser aleatorio o no, dependiendo del caso.

Si se decide imputar (“rellenar”) estos datos, podría introducirse un sesgo importante.

```python
#Revisión de columnas con missings
df.isnull().sum()
#df.info() #También indica los missings
```

```python
#Vemos cuánto es el porcentaje de missings de la columna Gender
(11/1000)*100
```

```python
df.shape
```

Las opciones más frecuentes de solucionar este problema:
- Eliminar columnas #Cuando los datos faltantes son muchos
- Eliminar filas (observaciones) #Cuando missings son pocos
- Imputar datos (por columna) #Cuando datos faltantes son relativamente pocos 

```python
df_sin_missings = df.dropna() #Elimina todas las filas con missings
df_sin_missings.info()
```

```python
#Cuando una columna tiene muchos missings, generalmente se elimna
df2 = df.drop(["Commute Distance", "Occupation"], axis=1) #Axis=1 indica que se elimina columna, axis=0 filas 
```

```python
#Visualizamos que se hayan eliminado las columnas
df2.head()
```

Imputación de missings:
- Dataframe.fillna(valor) #Se completa **todo el dataframe** con este valor
- Dataframe["columna"].fillna(valor) #Solo una columna

Estas son solo vistas, **para guardar cambios se debe sobreescribir**

```python
#Como ejercicio, imputaremos los missing values de Gender, Income y Childre
#Gender
moda_gender = df["Gender"].mode()
moda_gender #Se obtiene un array, una especie de lista
```

```python
moda_gender[0]
```

```python
df["Gender"] = df["Gender"].fillna("Male") #
```

```python
df.info()
```

```python
#PAra guardar se sobreescribe
df["Gender"] = df["Gender"].fillna(moda_gender[0])
```

```python
#Income con promedio
promedio_income = df["Income"].mean()
promedio_income
```

```python
df["Income"] = df["Income"].fillna(56267.60) #
df["Income"] = df["Income"].fillna(promedio_income) #
```

```python
#Children con mediana
mediana_children = df["Children"].median()
mediana_children
```

```python
df["Children"] = df["Children"].fillna(mediana_children)
```

# 5 Outliers
Dato atípico es una observación que se desvía tanto del resto de las observaciones que hace sospechar que ha sido generada por un mecanismo distinto.

Se pueden detectar:
- Estadísticas
- Gráficamente



```python
#  Visualizamos estadísticas nuevamente y vemos los mínimos y máximos
df.describe()
```

En estas variables, ¿Se observa algo atípico?

Gráficamente <br>

Box plot

```python
df.boxplot(column=["Children"])
```

```python
#Se puede juntar con más columnas.
df.boxplot(column=["Children", "Cars"])
```

```python
df.Cars.value_counts()
```

¿Cómo quitar outliers? Simplemente filtrando los valores deseados. En el caso anterior, si consideramos que tener 4 autos es un dato atípico, lo filtramos.

Filtrando se quitan outliers

```python
#df[df["Cars"]<=3]
df_filtrado = df[df["Cars"]<=3] #Para guardar se sobreescribe
```

Si queremos crear una tabla solo con personas jóvenes, utilizamos los filtros, en este caso para la columna "Age".

```python
df_joven = df[df["Age"]<=40]
```

```python
df_joven
```

```python
df.plot.scatter(x="Age", y="Income")
```

En este scatter plot, vemos que la observación de la persona de 89 años es un dato atípico en nuestro dataset, así que la eliminaremos.

```python
df = df[df["Age"]<=85] #Dejamos en nuestro df los menores de 85 años
```
